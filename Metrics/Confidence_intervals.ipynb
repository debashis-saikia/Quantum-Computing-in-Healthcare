```python
import numpy as np
import matplotlib.pyplot as plt
```

# Grouped Bar Chart

```python
models = ["SVC", "RF", "KNN", "LR", "UUâ€ ", "HQNN"]
accuracy_mean = [0.4889, 0.9800, 0.8633, 0.4489, 0.8001, 0.9501]
accuracy_ci_lower = [0.4378, 0.9371, 0.7863, 0.3720, 0.7615, 0.9192]
accuracy_ci_upper = [0.5400, 1.0229, 0.9403, 0.5258, 0.8387, 0.9810]

yerr_lower = np.array(accuracy_mean) - np.array(accuracy_ci_lower)
yerr_upper = np.array(accuracy_ci_upper) - np.array(accuracy_mean)

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(models, accuracy_mean, yerr=[yerr_lower, yerr_upper],
              capsize=5, alpha=0.7, color="skyblue", edgecolor="black")

ax.set_ylim(0, 1.1)
ax.set_ylabel("Accuracy", fontsize=12)
ax.set_title("Model Accuracy with 95% CI", fontsize=14, fontweight="bold")
ax.grid(axis="y", linestyle="--", alpha=0.6)

for bar, mean in zip(bars, accuracy_mean):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
            f"{mean:.2f}", ha="center", fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
models = ["SVC", "RF", "KNN", "LR", "UUâ€ ", "HQNN"]
accuracy_mean = [0.6900, 0.7700, 0.4750, 0.5950, 0.9200, 0.8150]
accuracy_ci_lower = [0.6264, 0.7302, 0.3962, 0.5577, 0.8771, 0.7773]
accuracy_ci_upper = [0.7536, 0.8098, 0.5538, 0.6323, 0.9629, 0.8530]

yerr_lower = np.array(accuracy_mean) - np.array(accuracy_ci_lower)
yerr_upper = np.array(accuracy_ci_upper) - np.array(accuracy_mean)

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(models, accuracy_mean, yerr=[yerr_lower, yerr_upper],
              capsize=5, alpha=0.7, color="skyblue", edgecolor="black")

ax.set_ylim(0, 1.1)
ax.set_ylabel("Accuracy", fontsize=12)
ax.set_title("Model Accuracy with 95% CI", fontsize=14, fontweight="bold")
ax.grid(axis="y", linestyle="--", alpha=0.6)

for bar, mean in zip(bars, accuracy_mean):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
            f"{mean:.2f}", ha="center", fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```

# Forest Plot

```python
sorted_idx = np.argsort(accuracy_mean)
models_sorted = [models[i] for i in sorted_idx]
mean_sorted = [accuracy_mean[i] for i in sorted_idx]
lower_sorted = [accuracy_ci_lower[i] for i in sorted_idx]
upper_sorted = [accuracy_ci_upper[i] for i in sorted_idx]

yerr_lower_sorted = np.array(mean_sorted) - np.array(lower_sorted)
yerr_upper_sorted = np.array(upper_sorted) - np.array(mean_sorted)

fig, ax = plt.subplots(figsize=(7, 5))

ax.errorbar(mean_sorted, models_sorted, 
            xerr=[yerr_lower_sorted, yerr_upper_sorted],
            fmt='o', color='navy', ecolor='gray', elinewidth=2, capsize=5)

ax.set_xlim(0, 1.1)
ax.set_xlabel("Accuracy", fontsize=12)
ax.set_title("Forest Plot of Model Accuracy with 95% CI", fontsize=14, fontweight="bold")
ax.grid(axis="x", linestyle="--", alpha=0.6)

for i, (mean, low, high) in enumerate(zip(mean_sorted, lower_sorted, upper_sorted)):
    ax.text(high + 0.01, i, f"[{low:.2f}, {high:.2f}]", va="center", fontsize=9)

plt.tight_layout()
plt.show()
```
